```python
import pandas as pd
import seaborn as sns

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Show first 5 rows
titanic.head()
```

```python
titanic.info()          # see columns, data types, missing values
print("Number of passengers:", len(titanic))
print("Average age:", titanic['age'].mean())
print("Survival rate:", titanic['survived'].mean() * 100, "%")
```

```python
titanic['class'].value_counts().plot(kind='bar', title='Passenger Classes')
```
